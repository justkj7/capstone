## Experiment
Define necessary methods for prediction.

```python
import tensorflow.compat.v1 as tf
import os 
import shutil
import csv
import pandas as pd
import IPython

tf.get_logger().setLevel('ERROR')

from tapas.utils import tf_example_utils
from tapas.protos import interaction_pb2
from tapas.utils import number_annotation_utils
from tapas.scripts import prediction_utils
from collections import OrderedDict

max_seq_length = 512
vocab_file = "tapas_sqa_base/vocab.txt"
```

```python
os.makedirs('results/sqa/tf_examples', exist_ok=True)
os.makedirs('results/sqa/model', exist_ok=True)
os.makedirs('results/sqa/opt', exist_ok=True)
with open('results/sqa/model/checkpoint', 'w') as f:
  f.write('model_checkpoint_path: "model.ckpt-0"')
for suffix in ['.data-00000-of-00001', '.index', '.meta']:
  shutil.copyfile(f'tapas_sqa_base/model.ckpt{suffix}', f'results/sqa/model/model.ckpt-0{suffix}')
```

```python
def write_tf_example(filename, examples_pair):
  token_cnt_list = []
  with tf.io.TFRecordWriter(filename) as writer:
    for pair in examples_pair:
      example = pair[0]
      token_cnt = pair[1]
      writer.write(example.SerializeToString())
      token_cnt_list.append(token_cnt)
  return token_cnt_list

def convert_interactions_to_examples(converter, tables_and_queries):
  """Calls Tapas converter to convert interaction to example."""
  for idx, (table, queries) in enumerate(tables_and_queries):
    interaction = interaction_pb2.Interaction()
    for position, query in enumerate(queries):
      question = interaction.questions.add()
      question.original_text = query
      question.id = f"{idx}-0_{position}"
    for header in table[0]:
      interaction.table.columns.add().text = header
    for line in table[1:]:
      row = interaction.table.rows.add()
      for cell in line:
        row.cells.add().text = cell if isinstance(cell, str) else ''
    number_annotation_utils.add_numeric_values(interaction)
    for i in range(len(interaction.questions)):
      try:
        yield converter.convert(interaction, i)
      except ValueError as e:
        print(f"Can't convert interaction: {interaction.id} error: {e}")

def predict(converter, table, queries):
  examples = convert_interactions_to_examples(converter, [(table, queries)])
  token_cnt_list = write_tf_example("results/sqa/tf_examples/test.tfrecord", examples)

  ! python -m tapas.run_task_main \
    --task="SQA" \
    --output_dir="results" \
    --noloop_predict \
    --test_batch_size={len(queries)} \
    --tapas_verbosity="ERROR" \
    --compression_type= \
    --init_checkpoint="tapas_sqa_base/model.ckpt" \
    --bert_config_file="tapas_sqa_base/bert_config.json" \
    --mode="predict" 2> error
    
  results_path = "results/sqa/model/test_sequence.tsv"
  all_coordinates = []
  df = pd.DataFrame(table[1:], columns=table[0])
  with open(results_path) as csvfile:
    reader = csv.DictReader(csvfile, delimiter='\t')
    for row in reader:
      coordinates = prediction_utils.parse_coordinates(row["answer_coordinates"])
      all_coordinates.append(coordinates)
      #answers = ', '.join([table[row + 1][col] for row, col in coordinates])
      #position = int(row['position'])
      #print(">", queries[position])
      #print(answers)
  return all_coordinates, token_cnt_list

def experiment_sqa(firstonly):
  for pruning_opt_level in range(0, 3):
    print()
    output_data = {}
    output_data['id'] = []
    output_data['annotator'] = []
    output_data['position'] = []
    output_data['token_size'] = []
    output_data['result'] = []
    
    config = tf_example_utils.ClassifierConversionConfig(
      vocab_file=vocab_file,
      max_seq_length=max_seq_length,
      max_column_id=max_seq_length,
      max_row_id=max_seq_length,
      strip_column_names=False,
      add_aggregation_candidates=False,
      pruning_opt_level=pruning_opt_level,
    )
    converter = tf_example_utils.ToClassifierTensorflowExample(config)

    #test_ds = pd.read_csv("SQA_dataset/test.tsv", delimiter='\t', header=0, nrows=10)
    test_ds = pd.read_csv("SQA_dataset/test.tsv", delimiter='\t', header=0)

    q_dict = OrderedDict()
    file_dict = dict()
    for idx, row in test_ds.iterrows():
      if not row['id'] in file_dict:
        file_dict[row['id']] = row['table_file']

      if not row['id'] in q_dict:
        q_dict[row['id']] = OrderedDict()
      if not row['annotator'] in q_dict[row['id']]:
        q_dict[row['id']][row['annotator']] = list()
        q_dict[row['id']][row['annotator']].append([]) # idx 0 for questions
        q_dict[row['id']][row['annotator']].append([]) # idx 1 for answer text
        q_dict[row['id']][row['annotator']].append([]) # idx 2 for answer coordinates

      if firstonly and int(row['position']) != 0:
        continue
      q_dict[row['id']][row['annotator']][0].append(row['question'])
      q_dict[row['id']][row['annotator']][1].append(row['answer_text'])
      q_dict[row['id']][row['annotator']][2].append(row['answer_coordinates'])

    for id, val in q_dict.items():
      # iteration by id
      for annotator, list_pair in val.items():
        # iteration by annotator
        table_ds = pd.read_csv("SQA_dataset/" + file_dict[id], delimiter=',', header=0, dtype='unicode')
        table = table_ds.values.tolist()
        table.insert(0, table_ds.columns.tolist())

        result_coordinates, token_cnt_list = predict(converter, table, list_pair[0])
        if result_coordinates == None:
          continue

        for pos, (result_coords, token_size) in enumerate(zip(result_coordinates, token_cnt_list)):
          answer_coordinates = prediction_utils.parse_coordinates(list_pair[2][pos])
          matched = result_coords == answer_coordinates
          print("opt(%i), id(%s), annotator(%i), pos(%i), token_size(%i), result(%r)" 
                  % (pruning_opt_level, id, annotator, pos, token_size, matched))
        
          output_data['id'].append(id)
          output_data['annotator'].append(annotator)
          output_data['position'].append(pos)
          output_data['token_size'].append(token_size)
          output_data['result'].append(matched)
    output_df = pd.DataFrame(output_data, columns=['id', 'annotator', 'position', 'token_size', 'result'])
    print(output_df)
    output_df.to_csv("results/sqa/opt/sqa_experiment_opt_%i_%s.csv" 
                     % (pruning_opt_level, ('firstonly' if firstonly else 'cluster')), mode='w', index=False)
```

Initiate prediction for the case of whole testcases in question sequences and the case of single testcase in question sequences.

```python
experiment_sqa(True)
experiment_sqa(False)
```

Analyze the results.
